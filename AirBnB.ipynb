```python
# Step 2: Initial Data Assessment
print("Missing values before cleaning:")
print(df.isnull().sum().sort_values(ascending=False))

print("\nDataset overview:")
print(f"Total rows: {len(df)}")
print(f"Total columns: {len(df.columns)}")
```

```python
# Step 3: Handle Missing Values

# Create a copy for cleaning
df_clean = df.copy()

# Check percentage of missing values
missing_pct = (df_clean.isnull().sum() / len(df_clean)) * 100
print("Missing value percentage:")
print(missing_pct.sort_values(ascending=False).round(2))
```

```python
# Step 2: Handle 'consumer' column (58.90% missing)
# Strategy: Create "Unknown" category since over half is missing

df_clean['consumer'] = df_clean['consumer'].fillna('Unknown')

print("✓ 'consumer' column: Missing values filled with 'Unknown'")
print(f"  Value counts: {df_clean['consumer'].value_counts().to_dict()}")
```

```python
# Clean bathrooms column
if 'bathrooms' in df_clean.columns:
    df_clean['bathrooms'] = pd.to_numeric(df_clean['bathrooms'], errors='coerce')
    print(f"✓ bathrooms: converted to float")

```

```python
# Create flag for imputed values
df_clean['reply_time_imputed'] = df_clean['reply time'].isnull()

# Ensure host response rate is properly formatted (0-1 scale)
if 'host response rate' in df_clean.columns:
    df_clean['host response rate'] = pd.to_numeric(df_clean['host response rate'], errors='coerce')

    # Create response rate categories for group-based imputation
    df_clean['host response rate'] = pd.cut(
        df_clean['host response rate'],
        bins=[0, 0.5, 0.9, 1.0],
        labels=['Low', 'Medium', 'High'],
        include_lowest=True
    )

    # Group-based imputation: fill missing values with median of each category
    category_medians = df_clean.groupby('response_rate_category')['reply time'].transform('median')
    df_clean['reply time'] = df_clean['reply time'].fillna(category_medians)

    # For any remaining missing values (if categories had no data), use overall median
    overall_median = df_clean['reply time'].median()
    df_clean['reply time'] = df_clean['reply time'].fillna(overall_median)

    print("✓ Reply time imputation completed using response rate categories")

print(f"Missing values before imputation: {df['reply time'].isnull().sum()}")
print(f"Missing values after imputation: {df_clean['reply time'].isnull().sum()}")
print(f"Total values imputed: {df_clean['reply_time_imputed'].sum()}")
```
